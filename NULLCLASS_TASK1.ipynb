```python
import numpy as np
import PyPDF2
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import sent_tokenize
import nltk

# Download punkt tokenizer if not already present
nltk.download('punkt')

def extractive_summary(text, num_sentences=3):
    # 1. Split text into sentences
    sentences = sent_tokenize(text)

    if len(sentences) <= num_sentences:
        return text  # if text is too short, return as is

    # 2. Convert sentences to TF-IDF vectors
    vectorizer = TfidfVectorizer()
    sentence_vectors = vectorizer.fit_transform(sentences)

    # 3. Compute sentence similarity matrix
    similarity_matrix = cosine_similarity(sentence_vectors)

    # 4. Score sentences
    scores = similarity_matrix.sum(axis=1)

    # 5. Get top N sentences
    top_indices = np.argsort(scores)[-num_sentences:]
    top_indices.sort()  # maintain original order

    # 6. Form summary
    summary = ' '.join([sentences[i] for i in top_indices])
    return summary


# ----------------------------
# Main program
# ----------------------------
choice = input("Do you want to load from a file? (y/n): ").strip().lower()

if choice == 'y':
    file_path = input("Enter the path to your .txt or .pdf file: ").strip()

    if file_path.lower().endswith(".pdf"):
        # Read PDF
        with open(file_path, "rb") as pdf_file:
            reader = PyPDF2.PdfReader(pdf_file)
            text_data = ""
            for page in reader.pages:
                text_data += page.extract_text() + "\n"
    else:
        # Read text file
        with open(file_path, "r", encoding="utf-8") as f:
            text_data = f.read()

else:
    print("\nPaste your large paragraph or document text (press Enter twice to finish):")
    lines = []
    while True:
        line = input()
        if line == "":
            break
        lines.append(line)
    text_data = " ".join(lines)

num_sentences = int(input("\nHow many sentences do you want in the summary? "))

summary = extractive_summary(text_data, num_sentences=num_sentences)

print("\n----- SUMMARY -----")
print(summary)

```
